```python
import pandas as pd
import numpy as np
import pickle
import json
```

```python
#importing the dictionary:
with open('dictionary.json') as json_file:
    dictionary = json.load(json_file)

dictionary 
```

```python
#importing data:
with open('data_tokenized.pickle', 'rb') as handle:
    tokenized_data = pickle.load(handle)
    
tokenized_data
```

```python
total = 0
i = 0
tweet_numberized = list()
for tweet in tokenized_data:
    aux = list()
    for word in tweet[1]:
        if(word in dictionary):
            aux.append(dictionary[word])
    if(len(aux)>2):#We only take vectors with 3 or more words in our dictionary
        tweet_numberized.append([np.array(aux), tweet[2]])
        total += 1
        i += len(aux)

        
print('Total of tweets: {}'.format(total))
print('Average of words/tweet: {}'.format(i/total))

        

colunms = ['X', 'y']
df_tweet_numberized = pd.DataFrame(tweet_numberized, columns=colunms)
df_tweet_numberized
```

```python
with open('dataset.pickle', 'wb') as handle:
    pickle.dump(df_tweet_numberized, handle)
```
