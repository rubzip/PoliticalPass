```python
import pandas as pd
import numpy as np
import pickle
import json
```

```python
#First step is to load the 
df_dictionary = pd.read_csv('df_freq.csv')

#We are going to take the n_max most frequent words:
n_max = 2000

df_dictionary = df_dictionary[:n_max]
df_dictionary
```

```python
#We are going to make a dictionary using Word as a key and index as values
df_dictionary = df_dictionary.reset_index()
df_dictionary = df_dictionary.set_index(keys='Word')
aux = df_dictionary['index'].to_dict()

#THIS IS THE MOST IMPORTANT FILE IN PROGRAM ---> 'dictionary.json'
with open('dictionary.json', 'w') as f:
    json.dump(aux, f)
```
