# Lemmatizing And Tokenizer

In this notebook I'm going to lemmatize tweets and make a tokenizer using Train-Test Data.

```python
import pandas as pd
import numpy as np
from datetime import datetime
import json
import io
from fun import * 

import spacy
#From spacy, we are going to use 'es_dep_news_trf'
nlp = spacy.load('es_dep_news_trf')

#Wordclouds:
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.text import Tokenizer
```

## Lemmatizing Train-Test Dataset

```python
df_train = pd.read_csv('../0_DATA/train-test_tweets.csv')
#This can solve us some headaches:
df_train = df_train.dropna()

df_train
```

```python
# Example of how the function lemmatize_tweet works:
example_tweet = df_train.loc[1, 'Tweet']

print('Original tweet:')
print(example_tweet)
print('\n\nLemmatized tweet:')
print(lemmatize_tweet(nlp, example_tweet))
```

```python
tic = datetime.now() 
toc = lambda: (datetime.now() - tic).total_seconds()/60
new_tweets = df_train['Tweet'].map(lambda x: lemmatize_tweet(nlp, x))

print('Execution time: {} min'.format(toc()))
```

```python
df_train['Tokenized'] = new_tweets
df_train.loc[:,['User', 'Tokenized', 'Left']].to_csv('../0_DATA/train-test_lemma.csv', index=False)

df_train.loc[:,['User', 'Tokenized', 'Left']]
```

## Tokenizer

```python
# Creation and fitting of the Tokenizer:
tokenizer = Tokenizer(num_words=5000, oov_token='<OOV>')
tokenizer.fit_on_texts(new_tweets)

save_tokenizer(tokenizer)
```

```python
words_freq = dict(tokenizer.word_counts)
words_freq
```

```python
#To clarify us and analyse results we are going to plot data as a wordcloud:
wordcloud = WordCloud(width = 1000, height = 500).generate_from_frequencies(words_freq)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)

wordcloud.to_file('../4_IMAGES/wordcloud.png')
```
